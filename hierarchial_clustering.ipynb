```python
def find_min(data):
    minimum=float('inf')
    n=len(data)
    x=y=0
    for i in range(1,n):
        for j in range(0,i):
            ele=data.iloc[i,j]
            if ele<minimum:
                minimum=ele
                x=i
                y=j
    return x,y
```

```python

def merge_clusters(data):
    
    n = len(data)
    if(n<=1):
        return data
        
    print("Older matrix : ")
    print(data,"\n")
    x, y = find_min(data)
    x, y = min(x, y), max(x, y)
    print("Merging", data.columns[x], "and", data.columns[y])
    new_name = "(" + data.columns[x] + " , " + data.columns[y] + ")"
    a = data.columns.tolist()
    a[x] = new_name
    data.columns = a
    
    
    for i in range(n):
        if(i!=x):
            e1 = data.iloc[i, x]
            e2 = data.iloc[i, y]
            minimum = min(e1, e2)
            data.iloc[i, x] = minimum
            data.iloc[x, i] = minimum
        

    
    data = data.drop(data.columns[y], axis=1)
    data = data.drop(y)
    
    # Reset index to have row index in ascending order starting from 0
    data = data.reset_index(drop=True)

    print("New Matrix :")
    print(data,"\n")
    return data
```

```python
p=data.copy()
for i in range(n+1):
    p=merge_clusters(p)
```
